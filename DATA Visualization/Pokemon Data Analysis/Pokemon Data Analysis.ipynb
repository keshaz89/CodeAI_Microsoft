```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(7,4))
sns.histplot(data=df, x='Total', bins=20, kde=True)
plt.title('Distribution of Total Stats')
plt.xlabel('Total Stats')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(7,4))
sns.scatterplot(data=df, x='Attack', y='Defense',hue='Legendary')
plt.title('Attack vs Defense')
plt.xlabel('Attack')
plt.ylabel('Defense')
plt.show()
```

```python
plt.figure(figsize=(7,4))
sns.countplot(data=df, x='Type 1', order=df['Type 1'].value_counts().index, palette='viridis')
plt.title('Frequency of Pokemon Type (Type 1)')
plt.xlabel('Type 1')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(7,4))
sns.boxplot(data=df,x='Generation',y='Total',palette='coolwarm')
plt.title('Total Stats Distribution by Generation')
plt.xlabel('Generation')
plt.ylabel('Total Stats')
plt.show()
```

```python
legendary_counts=df['Legendary'].value_counts()
plt.figure(figsize=(7,5))
plt.pie(legendary_counts, labels=legendary_counts.index, autopct='%.1f%%', colors=['skyblue', 'green'])
plt.title('Proportion of Legendary Pokemon')
plt.show()
```

```python
sns.set_palette('pastel')
type1_counts = df['Type 1'].value_counts()
plt.figure(figsize=(7,5))
plt.pie(type1_counts,labels=type1_counts.index,startangle=90,counterclock=False,autopct='%1.1f%%')
plt.title('Distribution of Type 1 Pokemon')
plt.axis('equal')
plt.show()
```

```python
generation_totals=df.groupby('Generation')['Total'].mean().reset_index()
plt.figure(figsize=(7,4))
sns.lineplot(data=generation_totals,x='Generation',y='Total',marker='o')
plt.title('Mean Total Stats over Generation')
plt.xlabel('Generation')
plt.ylabel('Mean Total Stats')
plt.xticks(generation_totals['Generation'])
plt.grid(True)
plt.show()
```

```python
numerical_columns = df[['Total', 'HP', 'Attack', 'Defense', 'Sp. Def', 'Speed', 'Generation']]
correlation_matrix = numerical_columns.corr()
plt.figure(figsize=(7,5))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix of Pokemon Stats')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig=plt.figure(figsize=(7,4))
ax=fig.add_subplot(111,projection='3d')
x=df['Attack']
y=df['Defense']
z=df['Total']
colors=df['Generation']
sc=ax.scatter(x,y,z,c=colors,cmap='viridis',s=50)
ax.set_xlabel('Attack')
ax.set_ylabel('Defense')
ax.set_zlabel('Total')
ax.set_title('3D scatter plot of attack,defense and total stats')
plt.colorbar(sc,label='Generation')
plt.show()
```

```python
columns_to_plot = ['Total', 'HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed', 'Generation', 'Legendary']
sns.pairplot(df[columns_to_plot], hue='Legendary')
plt.suptitle('Pair Plot of Pokemon Stats with Legendary Status')
plt.show()
```

```python
plt.figure(figsize=(7,4))
sns.violinplot(x='Legendary',y='Speed',data=df,inner='quartile',palette='muted')
plt.title('Violin Plot of Speed by Legendary Status')
plt.xlabel('Legendary')
plt.ylabel('Speed')
plt.xticks(ticks=[0,1],labels=['Not Legendary','Legendary'])
plt.show()
```

```python
df['Total without HP']=df['Total']-df['HP']
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
df[['Attack_scaled','Defense_scaled','Speed_scaled']]=scaler.fit_transform(df[['Attack','Defense','Speed']])
df.head()
```

```python
bins=[0,50,100,150,200]
labels=['very slow','slow','fast','very fast']
df['Speed_category']=pd.cut(df['Speed'],bins=bins,labels=labels)
df.head()
```

```python
fast_pokemons=df[df['Speed_category']=='fast']
print(fast_pokemons)
```
